```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("adult.data" , names = ["age" , "workclass" , "fnlwgt" , "education" , "education-num" , "marital-status" ,
                                     "occupation" , "relationship" , "race" , "sex" , "capital-gain" , "capital-loss" , "hours-per-week" ,
                                     "native-country" , "income"])
data
```

```python
data["income"].unique()
```

# Race
## Capital Gain
According to this chart, we can see that Asian immigrants have more capital compared to other races. They have more than twice as much capital compared to Black people and Indians.
## Capital Loss
The lost capital of Asian immigrants is also much higher. An important point to note is that the lost capital of Indian immigrants is very minimal, even less than that of Black people.


```python
race_reich = data.groupby("race" , as_index = False).agg({"capital-gain" : "mean"}).sort_values("capital-gain" , ascending = False)
race_reich
```

```python
fig,ax = plt.subplots()
color = ["#0a2851" , "#104587" , "#104587" , "#1a77e1" ,"#4d95ea"]
plt.barh(race_reich["race"] , race_reich["capital-gain"] , color = color)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(fontsize = 8 , color = "gray")
plt.yticks(color = "gray")
plt.title("Capital Gain for Race" , fontsize = 10 , loc = "left" , color = "gray")
plt.xlabel("Capital Gain" , fontsize = 10 , loc = "left" , color = "gray")
plt.ylabel("Race" , fontsize = 10 , loc = "bottom" , color = "gray")
plt.show()
```

```python
race_arm = data.groupby("race" , as_index = False).agg({"capital-loss" : "mean"}).sort_values("capital-loss" , ascending = False)
race_arm
```

```python
fig,ax = plt.subplots()
color = ["#0a2851" , "#104587" , "#104587" , "#1a77e1" ,"#4d95ea"]
plt.barh(race_arm["race"] , race_arm["capital-loss"] , color = color)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(fontsize = 8 , color = "gray")
plt.yticks(color = "gray")
plt.title("Capital Loss for Race" , fontsize = 10 , loc = "left" , color = "gray")
plt.xlabel("Capital Loss" , fontsize = 10 , loc = "left" , color = "gray")
plt.ylabel("Race" , fontsize = 10 , loc = "bottom" , color = "gray")
plt.show()
```

# Hours per Week and Capital
On average, men work nearly six hours more per week than women, which is a natural occurrence. However, what is very attention-grabbing is the amount of capital owned by men, which is significantly more than double that of women.
The average loss of capital for men is significantly higher than for women, indicating that men have taken much more risk in their work.



```python
fig,ax = plt.subplots()
color = ["#f5448f" , "#9400d3"]
ax.bar(gp1["sex"] , gp1["hours-per-week"] , color = color)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(fontsize = 10 , color = "gray")
plt.yticks(color = "gray")
plt.title("Hours-per-week for Sex" , fontsize = 10 , loc = "left" , color = "gray")
plt.xlabel("Sex" , fontsize = 10 , loc = "left" , color = "gray")
plt.ylabel("Hours per Week" , fontsize = 10 , loc = "bottom" , color = "gray")
plt.show()
```

# Hours per Week and Capital for Marital Status
The amount of weekly work hours for individuals in this chart is completely natural and not noteworthy, as considering the average age of individuals, it is natural for widows to have the lowest amount of working hours. Single individuals also, given that they have less responsibility and are likely to be students with less time for work, it is natural for them to have fewer working hours.



```python
gp2 = data.groupby("marital-status",as_index=False).agg({"hours-per-week" : "mean",
                                                         "age" : "mean", "hours-per-week" : "mean"})
gp2 = gp2.sort_values("hours-per-week",ascending = False)
gp2
```
